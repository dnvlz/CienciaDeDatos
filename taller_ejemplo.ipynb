```python
df = pd.read_csv("datasets/NISPUF17.csv")
mother_edu = df['EDUC1']
categories = ["menos que preparatoria","preparatoria","mÃ¡s que preparatoria pero sin tÃ­tulo","tÃ­tulo"]

proporciones = dict()
for category in categories:
    proporciones[category] = 0
for i in mother_edu:
    proporciones[categories[i-1]] += 1   
for category in categories:
    proporciones[category] = proporciones[category]/len(mother_edu)
# llave: valor
#proporciones["llave"] = "valor"
#proporciones

proporciones
df
```

```python
from numpy import array

condition = (df['CBF_01']==1)|(df['CBF_01']==2)

df = df[["CBF_01","P_NUMFLU"]][condition].dropna()

# # necesitamos dos arrays
# # uno que indique nÃºmero de vacunas por cada bebÃ© que fue amamantado
# # uno que indique el nÃºmero de vacunas por cada bebÃ© que no fue amamantado

condition_si = df["CBF_01"]==1
condition_no = df["CBF_01"]==2

dosis_si = array(df[condition_si]["P_NUMFLU"].mean())
dosis_no = array(df[condition_no]["P_NUMFLU"].mean())

print("El nÃºmero promedio de vacunas que recibieron los bebÃ©s amamantados fue de",dosis_si)
print("El nÃºmero promedio de vacunas que recibieron los bebÃ©s no amamantados fue de",dosis_no)
```
