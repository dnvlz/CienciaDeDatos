# Archivos CSV

<br>
mpg.csv, gasto de combustible para 234 carros.

* mpg : millas por galón
* class : clasificación de autos
* cty : millas por galón (ciudad)
* cyl : no. de cilindros
* displ : desplazamiento del motor en litros
* drv : f = tracción delantera, r = tracción trasera, 4 = 4wd
* fl : combustile (e = etanol E85, d = diesel, r = regular, p = premium, c = CNG)
* hwy : nillas por galón (carretera)
* manufacturer : fabricante
* model : modelo
* trans : transmisión
* year : año

```python
import csv

%precision 2

# abrimos archivo y convertimos a lista de diccionarios
with open('mpg.csv') as csvfile:
    mpg = list(csv.DictReader(csvfile))
    
mpg[:4]
```

```python
len(mpg)
```

```python
# columnas de doc .csv
mpg[0].keys()
```

```python
# Gasto de combustible promedio (ciudad).
sum(float(d['cty']) for d in mpg) / len(mpg)
```

```python
# Gasto promedio en carretera
sum(float(d['hwy']) for d in mpg) / len(mpg)
```

```python
# set: conjunto
cylinders = set(d['cyl'] for d in mpg)
cylinders
```

```python

# Ejemplo más complejo en el que agrupamos los autos por número de cilindros y encontramos el gasto en ciudad promedio para cada grupo.
CtyMpgByCyl = []

for c in cylinders: # iteramos sobre todos los cilindros
    summpg = 0
    cyltypecount = 0
    for d in mpg: # iteramos sobre todos los diccionarios
        if d['cyl'] == c: # si el tipo de cilindro coincide con el del diccionario actual
            summpg += float(d['cty']) # lo agregamos a la suma
            cyltypecount += 1 # incrementamos la cuenta
    CtyMpgByCyl.append((c, summpg / cyltypecount)) # tupla cilindro, promedio ('cylinder', 'avg mpg')

CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
# conjunto: valores únicos para la columna class
vehicleclass = set(d['class'] for d in mpg) # what are the class types
vehicleclass
```

```python
# Ejemplo de cómo encontrar el gasto promedio en carretera para cada clase de vehículo en nuestro conjunto de datos.
HwyMpgByClass = []

for t in vehicleclass: # iteamos sobre todas las clases de vehículo
    summpg = 0
    vclasscount = 0
    for d in mpg: # iteramos sobre diccionarios
        if d['class'] == t: # si coincide con clase de diccionario
            summpg += float(d['hwy']) # sumamos
            vclasscount += 1 # incrementamos la suma
    HwyMpgByClass.append((t, summpg / vclasscount)) # agregamos tupla clase, promedio ('class', 'avg mpg')

HwyMpgByClass.sort(key=lambda x: x[1])
HwyMpgByClass
```

#  Objetos y función map()

```python
# clase en python
class Person:
    department = 'Facultad de ciencias' # variable de clase

    def set_name(self, new_name): # método
        self.name = new_name
    def set_location(self, new_location):
        self.location = new_location
```

```python
person = Person()
person.set_name('Daniel')
person.set_location('CDMX')
print('{} vive en {} y estudió en {}'.format(person.name, person.location, person.department))
```

```python
# función min
store1 = [10.00, 11.00, 12.34, 2.34]
store2 = [9.00, 11.10, 12.34, 2.01]
cheapest = map(min, store1, store2)
cheapest
```

```python
# map: evaluación floja. Debemos iterar
for item in cheapest:
    print(item)
```

# Lambda y "List Comprehensions"

```python
# lambda: funciones rápidas
my_function = lambda a, b, c : a + b
```

```python
my_function(1, 2, 3)
```

```python
# iterar de 0 a 999 y regresar números pares
my_list = []
for number in range(0, 1000):
    if number % 2 == 0:
        my_list.append(number)
my_list
```

```python
# lo mismo con list comprehension
my_list = [number for number in range(0,1000) if number % 2 == 0]
my_list
```
